# Key EDA Findings

### 1. NAV Growth Trends

Daily NAV analysis showed consistent long-term growth across most mutual fund schemes between 2022 and 2026, despite periodic market fluctuations.
**Reference:** NAV Trend Analysis (Plotly Line Chart)


### 2. Impact of Market Cycles

Several mutual fund schemes demonstrated stronger NAV appreciation during the 2023 market rally, while periods of volatility in 2024 resulted in temporary slowdowns and corrections across selected funds.

### 3. Strong Growth in SIP Participation

Monthly SIP inflows increased steadily throughout the analysis period, reaching an all-time high of ₹31,002 Cr in December 2025.
**Reference:** SIP Inflow Time-Series Chart

### 4. Expanding Retail Investor Base

Mutual fund folio counts nearly doubled from 13.26 crore in January 2022 to 26.12 crore by late 2025, indicating significant retail participation growth.
**Reference:** Folio Count Growth Line Chart

### 5. Age Group Participation Patterns

Investors in the working-age categories contributed the largest share of mutual fund participation, reflecting higher investment activity among salaried individuals.
**Reference:** Age Group Distribution Pie Chart

### 6. SIP Contribution Varies by Age Group

Box plot analysis revealed noticeable differences in SIP investment amounts across age groups, with certain segments demonstrating higher median investment values.
**Reference:** SIP Amount by Age Group Box Plot

### 7. Gender Distribution Insights

The investor base showed a measurable gender imbalance, highlighting differences in participation levels across genders.
**Reference:** Gender Split Pie Chart

### 8. Geographic Concentration of Investments

A significant portion of SIP investments originated from a limited number of states, indicating regional concentration in mutual fund adoption.
**Reference:** State-wise SIP Amount Horizontal Bar Chart

### 9. Growing Presence Beyond Metro Cities

Although T30 cities contributed the majority of SIP inflows, B30 cities represented a meaningful share, reflecting increasing mutual fund penetration in emerging markets.
**Reference:** T30 vs B30 City Tier Pie Chart

### 10. Diversification Across Funds and Sectors

Correlation analysis revealed low pairwise correlations among the selected mutual funds, indicating relatively independent return movements and potential diversification benefits. Sector allocation analysis further demonstrated exposure across multiple industries, reducing concentration risk within the overall portfolio.
